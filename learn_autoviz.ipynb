### Import iris dataset

```python
from sklearn.datasets import load_iris 

data = load_iris()
```

```python
import numpy as np

full_data = np.append(data.data, data.target.reshape(-1,1), axis=1)
```

```python
import pandas as pd 

df = pd.DataFrame(full_data)
df.columns = data.feature_names + ['Type']
```

```python
df
```

```python
df.to_csv('iris.csv')
```
